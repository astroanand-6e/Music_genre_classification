# MERT-v1-95M Baseline Results on FMA-Small

## Progress Report: Music Genre Classification

This notebook documents the baseline performance of MERT-v1-95M with a linear probe on the FMA-Small dataset.

## Dataset Summary

- **Dataset**: FMA-Small (8,000 tracks, 30 seconds each)
- **Genres**: 8 balanced genres
- **Split**: 80% train (6,393 tracks), 20% test (1,599 tracks)
- **Features Extracted**: 7,994 tracks successfully
- **Failed Tracks**: 6 (corrupted MP3)

## Model & Method

- **Model**: MERT-v1-95M (95 million parameters)
- **Feature Extraction**: Mean-pooling of last hidden state over 10s audio clips
- **Classifier**: Logistic Regression (linear probe)
- **Device**: NVIDIA RTX 5060 Ti (16 GB VRAM)
- **Sampling Rate**: 24 kHz (MERT target rate)

## Results

This is a strong baseline for an unsupervised pre-trained model without fine-tuning.

### Key Observations

#### Strengths ✅
1. **Hip-Hop** (69% F1) — Distinctive rhythmic patterns and production techniques
2. **Folk** (66% F1) — Acoustic instruments and traditional structures
3. **International** (66% F1) — Cultural-specific musical elements
4. **Rock** (62% F1) — Strong instrumental signatures

#### Challenges ⚠️
1. **Pop** (32% F1) — Highly heterogeneous, borrows from other genres
2. **Experimental** (53% F1) — Ambiguous boundaries, diverse sub-styles
3. **Electronic/Instrumental** overlap — Both minimize vocal content

#### Model Insights
- MERT captures **rhythmic and timbral** features well
- Struggles with **genre fusion** and **crossover tracks**
- Strong performance on genres with distinct acoustic fingerprints

## Technical Details

### Feature Extraction Pipeline
1. **Audio Loading**: MP3 → waveform (torchaudio/librosa fallback)
2. **Preprocessing**: Resample to 24kHz, convert to mono
3. **Segmentation**: Extract first 10 seconds
4. **MERT Processing**: 768-dim embeddings via mean-pooling
5. **Classification**: Logistic Regression with L2 regularization

### Performance Metrics
- **Extraction Success Rate**: 99.93% (7,994/8,000)
- **Processing Time**: ~12 minutes on RTX 5060 Ti
- **Memory Usage**: ~900 MB GPU VRAM
